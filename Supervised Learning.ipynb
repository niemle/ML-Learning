```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
import numpy as np
import seaborn as sns

# Load Iris dataset
iris_bunch = datasets.load_iris()

# Convert to DataFrame
iris = pd.DataFrame(data=iris_bunch.data, columns=iris_bunch.feature_names)

# Add species names (target) as a new column
iris['species'] = iris_bunch.target_names[iris_bunch.target]

# Scatterplot with hue = species
plt.figure(num =1)
sns.scatterplot(data=iris, x='sepal length (cm)', y='sepal width (cm)', hue='species')
plt.title('Sepal')

plt.figure(num=2)
sns.scatterplot(data=iris, x='petal length (cm)', y='petal width (cm)', hue='species')
plt.title('Petal')
plt.show()

```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Needed for 3D projection (but no need to call directly)
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn import datasets

# Load Iris dataset
iris_bunch = datasets.load_iris()

# Convert to DataFrame
iris = pd.DataFrame(data=iris_bunch.data, columns=iris_bunch.feature_names)

# Add species names (target) as a new column
iris['species'] = iris_bunch.target_names[iris_bunch.target]

# Extract species numeric labels for coloring
species = iris_bunch.target

# PCA model for explained variance
pca_model = PCA(n_components=4)
pca_model.fit(iris_bunch.data)
explained_variance_ratio = pca_model.explained_variance_ratio_
var = np.cumsum(np.round(explained_variance_ratio, decimals=3)*100)

# PCA reduction to 3 components
train_reduced = PCA(n_components=3).fit_transform(iris_bunch.data)

# Create 3D scatter plot
fig = plt.figure(num=3, figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')  #  Correct way to add a 3D subplot

# Scatter plot of PCA-reduced data
sc = ax.scatter(train_reduced[:, 0], train_reduced[:, 1], train_reduced[:, 2], c=species, cmap='viridis', s=50)

# Labels and title
ax.set_title('Iris Dataset by PCA', size=14)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

# Colorbar for species
plt.colorbar(sc, ax=ax, label='Species')

# Show plot
plt.show()
```
